```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
    return np.sin(x)

def df(x):
    return np.cos(x)

dx = 0.1
```

```python
y_f = f(np.arange(0, 20, dx))
y_df = df(np.arange(0, 20, dx))
```

```python
plt.plot(np.arange(0, len(y_f)), y_f, c="red")
plt.plot(np.arange(0, len(y_df)), y_df, c="green")

plt.title("red - f(x); green - df(x)")
```

```python
y_udf = np.diff(y_f) / dx
```

```python
plt.plot(np.arange(0, len(y_udf)), y_udf, c="red")
plt.plot(np.arange(0, len(y_df)), y_df, c="green")
plt.plot(np.arange(0, len(y_f)), y_f, c="blue")

plt.grid()

plt.title("red - np.diff; green - df(x); blue - f(x)")
```

```python
print(y_df[:4])
print(y_udf[:4])
```

```python
y_differencial = y_udf * dx
```

```python
plt.plot(np.arange(0, len(y_differencial)), y_differencial, c="red")
plt.plot(np.arange(0, len(y_f)), y_f, c="green")

plt.title("red - df(x) * dx; green - f(x)")
```

```python
integral = 0
y_i = np.zeros(len(y_differencial) + 1)
for i, y in enumerate(y_differencial):
    integral += y
    y_i[i + 1] = integral 

print(y_f[:4])
print(y_differencial[:4])
print(y_i[:4])
```

```python
plt.plot(np.arange(0, len(y_f)), y_f, c="red")
plt.plot(np.arange(0, len(y_differencial)), y_differencial, c="green")
plt.plot(np.arange(0, len(y_i)), y_i, c="blue")

plt.title("red - f(x); green = df(x) * dx; blue - integral(df(x) * dx)")

plt.grid()
```

```python
def derrivative(i_x, dx):
    return np.diff(i_x)/dx

def parabola(x):
    return x**2 - x - 5
```

Геометрический смысл производной - нахождение секущей к графику, поскольку tg угла полученного треугольника между точками x0 ... x0 + dx и y0 ... y0 + dx равен (y0 + dy + y0) / (x0 + dx + x0) то тангенс угла секущей равен производной функции между данными точками. Далее можно подставить полученный угловой коэффициент в формулу прямой и получить секующую.

```python
p_x = np.arange(-10, 11, 1)
p_y = parabola(p_x)

dp_y = derrivative(p_y, 1)

x0 = len(p_x)//2
dx_p = 4
dx_x0 = x0 + dx_p

sek = ((p_y[dx_x0] - p_y[x0]) / dx_p) * p_x + p_y[x0]

plt.plot(p_x, p_y, marker=".")
plt.plot(p_x, sek)
plt.grid()
```

Следующие понятия дают предтсавление об анализе производной функции
df(x) = 0 - функция находится в минимуме или максимуме
df(x) < 0 - функция убывает
df(x) > 0 - функция возрастает 

Чтобы различать минимум и максимум необходимо посчитать производную второго порядка, производная второго порядка имеет следующие понятия
df(df(x)) = 0 - функция либо возрастает или убывает
df(df(x)) > 0 - функция находится в минимуме
df(df(x)) < 0 - функция находится в максимуме

Если иметь производную первого порядка и второго порядка можно легко определить минимум и максимум функции

```python
dx = 0.1

x_t = np.arange(0, 20, dx)
y_t = f(x_t)

d_yt = derrivative(y_t, dx)
d2_yt = derrivative(d_yt, dx)

plt.plot(x_t, y_t, c='red')
plt.plot(x_t[:-1], d_yt, c='green')
plt.plot(x_t[:-2], d2_yt, c='blue')

plt.title("red  - f(x); green - df(x); blue - d2f(x)")

plt.grid()
```
